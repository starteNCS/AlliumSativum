```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(10, 6))
# 1. Define the colormap
cmap = plt.get_cmap('viridis')

# 2. Define your labels and logic
labels = [
    'Reconstructed Worstcase',
    'Reconstructed BiModal',
    'Reconstructed UniModal',
    'Reconstructed Uniform',
    'Histogram'
]
base = 3 * 8
perMode = 3 * 8

data_x = [
    [1, 100],
    [3, 100],
    [3, 100],
    [1, 100],
    [1, 100]
]

data_y = [
    [base, base + 50 * perMode],
    [base + 2 * perMode, base + 2 * perMode],    
    [base + perMode, base + perMode],
    [base, base],
    [8, 100 * 8 * 2] # position and height
]

# 3. Plot using the colormap
for i, y in enumerate(data_y):
    color = cmap(i / (len(data_y) - 1)) 
    label=labels[i]
    plt.plot(data_x[i], y, label=label, color=color, linewidth=2)
    
plt.plot([6,6], [0, 1600], color='red', linestyle='--')
plt.fill_between(data_x[0], data_y[0], data_y[4], alpha=0.2)

plt.title('Storage need for histograms')
plt.xlabel('Bin count')
plt.ylabel('Size in byte')
plt.legend(loc='upper left')
plt.tight_layout()
plt.savefig('images/histogram/histogram_space_complexity.svg')
plt.show()
```
