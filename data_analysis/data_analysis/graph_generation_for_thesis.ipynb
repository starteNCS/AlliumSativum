```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

colors = plt.cm.viridis(np.linspace(0, 0.9, 2)) 
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

labels = ['Male', 'Female']
dataset = [1072, 1043]
assumed = [1072/2, 1043/2]
real = [81, 924]

x = np.arange(len(labels))  # the label locations

axes[0].bar(0, dataset[0], color=colors[0])
axes[0].bar(1, dataset[1], color=colors[1])
axes[0].set_ylim(0, 1100)
axes[1].bar(0, assumed[0], color=colors[0])
axes[1].bar(1, assumed[1], color=colors[1])
axes[1].set_ylim(0, 1100)
axes[2].bar(0, real[0], color=colors[0])
axes[2].bar(1, real[1], color=colors[1])
axes[2].set_ylim(0, 1100)

# Add some text for labels, title and custom x-axis tick labels, etc.
for i in range(len(labels) + 1):
    axes[i].set_xticks(x)
    axes[i].set_xticklabels(labels)
    
    
axes[0].set_ylabel('Count')
axes[0].set_title('Dataset')
axes[1].set_title('Delta Assumption')
axes[2].set_title('Filter executed')

plt.savefig('images/thesis/independence_assumption.svg', bbox_inches='tight', pad_inches=0.1)
plt.show()
plt.close()
```

```python
# FOR EQUALS


import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

colors = plt.cm.viridis(np.linspace(0, 0.9, 2)) 
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(20, 12))

dataset_x = [1, 2, 3, 4, 5]
dataset = [3, 5, 11, 7, 2]

ax1.bar(dataset_x, dataset, color=colors[0])

ax1.set_ylabel('Count')
ax1.set_xlabel('Key')
ax1.set_ylim(0, 12)
ax1.set_title('Base Dataset Histogram')


after_dataset = [0, 5, 0, 0, 0]

ax2.bar(dataset_x, after_dataset, color=colors[0])

ax2.set_ylabel('Count')
ax2.set_xlabel('Key')
ax2.set_ylim(0, 12)
ax2.set_title('After "=" Histogram')

plt.savefig('images/thesis/presentation/select_equal.svg', bbox_inches='tight', pad_inches=0.1)
plt.show()
plt.close()
```
